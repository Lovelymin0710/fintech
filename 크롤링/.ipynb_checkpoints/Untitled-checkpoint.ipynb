```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome("./chromedriver")
time.sleep(5)
url=f"https://www.hf.go.kr/hf/sub05/sub01.do?mode=list&srCategoryId1=&srSearchKey=&srSearchVal="
driver.get(url)
```

```python
element = driver.find_element("id", "search_val")
element.send_keys("결혼")
```

```python
import urllib.request as req

res=driver.page_source
soup=BeautifulSoup(res,"html.parser")
```

```python
print(soup)
```

```python
content_total=[]
link_total=[]
for i in range(1,40,1): 
  path=("#jwxe_main_content > div > div > div.board-list01.list-box > table > tbody > tr:nth-of-type({}) > td.title.left > a".format(i))
  content=soup.select(path)

  for i in content:
    content_text=i.get_text()
    content_text2=content_text.strip()
    content_total.append(content_text2)


  for i in content:
      link=i.attrs['href']  
      link_total.append("https://www.hf.go.kr/hf/sub05/sub01.do"+link)
```
